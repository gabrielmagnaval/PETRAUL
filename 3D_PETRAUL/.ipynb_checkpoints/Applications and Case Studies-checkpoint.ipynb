```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from matplotlib.backends.backend_tkagg import FigureCanvasTkAgg
import matplotlib.patches as patches
from matplotlib.patches import Ellipse
from matplotlib.lines import Line2D
import model
```

```python
def file_opening(name):
    """
    Open a CSV file and extract scenario names and corresponding data.

    Parameters:
    name (str): Path to the CSV file.

    Returns:
    tuple:
        - scenarios (list): List of scenario names (column headers excluding the first one).
        - data (list of arrays): List of arrays containing data for each scenario.
    """
    # Read the CSV file
    df = pd.read_csv(name, sep=';')

    # Extract scenario names
    scenarios = df.columns[1:].tolist()

    # Extract corresponding data as a list of arrays
    data = [df.iloc[:, k+1].values for k in range(len(scenarios))]

    return scenarios, data
```

```python
def file_opening_dict(name):
    with open(name) as file_name:
        df = pd.read_csv(file_name,index_col=0, sep=';')
        scenarios = df.to_dict(orient='list')
    return scenarios
```

```python
def file_opening_dict(file_name):
    """
    Open a CSV file and return its contents as a dictionary.

    Parameters:
    file_name (str): Path to the CSV file.

    Returns:
    dict: A dictionary where keys are column names and values are lists of column data.
    """
    # Read the CSV file with the first column as index
    df = pd.read_csv(file_name, index_col=0, sep=';')

    # Convert dataframe to dictionary with lists as values
    scenarios = df.to_dict(orient='list')

    return scenarios
```

# Application 1: sensitivity of contributors on EC

```python
### initialisation of the dictionnaries with the pre-set configurations.

contributor=['Body', 'Engine', 'Drivetrain', 'Path', 'Driver','Battery']

body_dict=file_opening_dict('body.csv')
engine_dict=file_opening_dict('thermal_ecoinvent.csv')
trans_dict=file_opening_dict('transmission.csv')
path_dict=file_opening_dict('path.csv')
driver_dict=file_opening_dict('driver.csv')

body_dict_EV=file_opening_dict('body_EV.csv')
engine_dict_EV=file_opening_dict('electric.csv')
trans_dict_EV=file_opening_dict('trans_EV.csv')
battery_dict_EV=file_opening_dict('battery.csv')

```

```python
### initialisation of the GV categories.

Reference=['Average_EU','Average_EU','Average_EU','EU_mix','Average_EU']

body_categories=['Mini','Compact','SUV','Pick-ups']
engine_categories=['Small_min','Small_EU','Large_EU','Large_max']
trans_categories=['Manual - FWD','Auto - FWD','Manual - AWD','Manual - RWD']
path_categories=['City - Mean','City – Dense','Hwy EU','Countryside']
driver_categories=['Extra Eco','Ecodriver','Aggres.','Extra Aggres.']

categories=[body_categories,engine_categories,trans_categories,path_categories,driver_categories]

Scenarios=[Reference]
for k in range (len(categories)):
    category=categories[k]
    for scenario in category:
        under_study=Reference.copy()
        under_study[k]=scenario
        Scenarios+=[under_study]

N_config=[len(body_categories),len(engine_categories),len(trans_categories),len(path_categories),len(driver_categories)]

labels_cat=['Reference']+body_categories+engine_categories+trans_categories+path_categories+driver_categories
```

```python
### initialisation of the BEV categories.

Reference_EV=['Average_EU','Average_EU','Average_EU','EU_mix','Average_EU','Average']

body_categories_EV=['Mini','Compact','SUV','Pick-ups']
engine_categories_EV=['Extra Small','Small','Large','Extra Large']
trans_categories_EV=['FWD','AWD','RWD']
path_categories_EV=['City - Mean','City – Dense','Hwy EU','Countryside']
driver_categories_EV=['Extra Eco','Ecodriver','Aggres.','Extra Aggres.']
battery_categories_EV=['Best','Worst']

categories_EV=[body_categories_EV,engine_categories_EV,trans_categories_EV,path_categories_EV,driver_categories_EV,battery_categories_EV]

Scenarios_EV=[Reference_EV]
for k in range (len(categories_EV)):
    category=categories_EV[k]
    for scenario in category:
        under_study=Reference_EV.copy()
        under_study[k]=scenario
        Scenarios_EV+=[under_study]

N_config_EV=[len(body_categories_EV),len(engine_categories_EV),len(trans_categories_EV),len(path_categories_EV),len(driver_categories_EV),len(battery_categories_EV)]

labels_cat_EV=['Reference']+body_categories_EV+engine_categories_EV+trans_categories_EV+path_categories_EV+driver_categories_EV+battery_categories_EV
```

```python
def EC_categories(scenario_input):
    """
    Computes energy consumption for multiple vehicle configurations.

    Parameters:
    - scenario_input (list of lists): Each sublist contains the pre-set configurations names analysed for the main contributors:
        [body, engine, transmission, path, driver]

    Returns:
    - ECs (list): Computed energy consumption values for each configuration.
    - Max (float): Maximum energy consumption value among all configurations studied.
    """
    ECs = []
    max_values = []

    for category_set in scenario_input:
        
        body_category, engine_category, trans_category, path_category, driver_category = category_set
        
        # Retrieve data from dictionaries
        body = body_dict[body_category]
        engine = engine_dict[engine_category]
        path = path_dict[path_category]
        trans = trans_dict[trans_category]
        driver = driver_dict[driver_category]
        
        # Compute energy consumption
        EC_value = model.EC_th(body, engine, trans, path, driver)
        ECs.append(EC_value)
        max_values.append(EC_value[0])  # Store only the primary EC value for max computation

    # Compute the maximum EC value
    max_ec = np.max(max_values)

    return ECs, max_ec
```

```python
def EC_categories_EV(scenario_input_EV):
    """
    Computes energy consumption for multiple vehicle configurations.

    Parameters:
    - categories (list of lists): Each sublist contains the pre-set configurations names analysed for the main contributors:
        [body, engine, transmission, path, driver]

    Returns:
    - ECs (list): Computed energy consumption values for each configuration.
    - Max (float): Maximum energy consumption value among all configurations studied.
    """
    ECs = []
    max_values = []

    for category_set in scenario_input_EV:
        body_category, engine_category, trans_category, path_category, driver_category, battery_category = category_set
        
        # Retrieve data from dictionaries
        body = body_dict_EV[body_category]
        engine = engine_dict_EV[engine_category]
        path = path_dict[path_category]
        trans = trans_dict_EV[trans_category]
        driver = driver_dict[driver_category]
        battery= battery_dict_EV[battery_category]
        
        # Compute energy consumption
        EC_value = model.EC_el(body, engine, trans, path, driver,battery)
        ECs.append(EC_value)
        max_values.append(EC_value[0])  # Store only the primary EC value for max computation

    # Compute the maximum EC value
    max_ec = np.max(max_values)

    return ECs, max_ec
```

```python
def sensitivity_contributor(ECs, N, m, contributor,EV=False):
    """
    Plots a bar chart comparing energy consumption for the different pre-set configurations of a given contributor.

    Parameters:
    - ECs (list of lists): Energy consumption values, where ECs[k][0] represents the energy consumption for configuration k.
    - N_config (list): Number of configurations for each category.
    - N (int): Index of the category to plot.
    - m (float): Maximum y-axis value for scaling.
    - contributor (str): contributor name for the label.
    """
    if EV:
        categories_name=labels_cat_EV
        config=N_config_EV
    else:
        categories_name=labels_cat
        config=N_config
    
    # Ensure valid index N
    if N >= len(config):
        print("Error: N is out of range for N_config")
        return
    
    # Compute indices for the category to plot
    step = np.cumsum(np.array(config))
    step = np.concatenate(([0], step))

    colors = ['red', 'blue', 'purple', 'green', 'orange','gray']
    
    # Reference value (baseline)
    ref = round(ECs[0][0], 2)
    ref_label = categories_name[0]
    L = [[ref, ref_label, 'black']]  # Baseline in black

    # Extract values for the selected category
    for k in range(step[N], step[N + 1]):
        EC = round(ECs[k + 1][0], 2)
        label = categories_name[k + 1]
        color = colors[N]
        L.append([EC, label, color])

    # Sort the list by energy consumption values
    L.sort(key=lambda x: x[0])

    # Create the plot
    
    label_order = [item[1] for item in L]
    values = [item[0] for item in L]
    bar_colors = [item[2] for item in L]

    plt.bar(range(len(L)), values, color=bar_colors)
    
    # Annotate bars with values
    for i, value in enumerate(values):
        if EV:
            plt.text(i, value + 1, f'{value}', ha='center', va='center', fontsize=14)
        else:
            plt.text(i, value + 0.3, f'{value}', ha='center', va='center', fontsize=14)

    # Formatting
    plt.ylabel(f'Contribution {contributor} [L/100km]', fontsize=14)
    plt.xticks(range(len(label_order)), label_order, fontsize=14, rotation=45)  # Rotate for readability
    plt.ylim(0, m * 1.1)

    # Show the plot
    plt.show()
```

```python
EV=False
if EV:
    ECs,Max=EC_categories_EV(Scenarios_EV)
else:
    ECs,Max=EC_categories(Scenarios)
k=3
sensitivity_contributor(ECs,k,Max,contributor[k],EV)
```

# Application 2 - Comparison of the parametrized model with ecoinvent

```python
### initialisation of dictionnaries with specific configurations representing ecoinvent model ###

body_dict=file_opening_dict('body.csv')
engine_dict=file_opening_dict('thermal_ecoinvent.csv')
trans_dict=file_opening_dict('transmission.csv')
path_dict=file_opening_dict('path_ecoinvent.csv')
driver_dict=file_opening_dict('driver.csv')

body_dict_EV=file_opening_dict('body_EV.csv')
engine_dict_EV=file_opening_dict('electric.csv')
trans_dict_EV=file_opening_dict('trans_EV.csv')
battery_dict_EV=file_opening_dict('battery.csv')
```

```python
def comparison_uncertainty_box(EC_p, EC_min, EC_max, EC_av_EU, EC_av_US):
    """
    Compares Ecoinvent energy consumption with the parametrized model across the 4 vehicle categories from Ecoinvent.

    Parameters:
    - title (str): Title of the plot.
    - EC_p (list): Principal EC values (main computed values).
    - EC_min (list of lists): Minimum EC values per category.
    - EC_max (list of lists): Maximum EC values per category.
    - EC_av_EU (list of lists): Average EC values for Europe.
    - EC_av_US (list of lists): Average EC values for the US.
    """
    
    fig, ax = plt.subplots(figsize=(8, 5))
    
    # Vehicle categories
    labels = ['Small', 'Medium', 'Large', 'Electric']
    
    x_positions = np.arange(len(EC_p))
    ymax = []

    # Plotting uncertainty ranges and values
    for j in range(len(EC_p)):
        principal = EC_p[j]
        minimum = EC_min[j]
        average_EU = EC_av_EU[j]
        average_US = EC_av_US[j]
        maximum = EC_max[j]

        # Plot average EU & US EC for GV
        ax.plot([j - 0.4, j + 0.4], [average_EU, average_EU], 'b-', label='EU Average' if j == 0 else "")
        ax.plot([j - 0.4, j + 0.4], [average_US, average_US], 'r-', label='US Average' if j == 0 else "")

        # Box for EC uncertainty range
        rect = plt.Rectangle((j - 0.4, minimum), 0.8, maximum - minimum, facecolor='lightgrey', edgecolor='black', linewidth=1, label='Best-Worst Case Range' if j == 0 else "")
        ax.add_patch(rect)

        # Principal ecoinvent value
        ax.scatter(j, principal, color='green', label='Principal EC' if j == 0 else "")

        ymax.append(maximum)

    # Plotting the second axes with kWh unit for BEV
    ax2 = ax.twinx()

    # Adjust y-axis limits
    max_y = np.max(ymax) * 1.1
    ax.set_ylim(0, max_y)
    
    conv_factor = 323 / 36  # Conversion factor to kWh/100km
    ax2.set_ylim(0, max_y * conv_factor)

    # Set x-axis labels
    ax.set_xticks(np.arange(len(EC_p)))
    ax.set_xticklabels(labels, fontsize=12)

    # Labels and legend
    ax.set_ylabel('Energy Consumption [L/100km]', fontsize=12)
    ax2.set_ylabel('Energy Consumption [kWh/100km]', fontsize=12)
    ax.legend(loc='upper right', bbox_to_anchor=(1, 1.2), ncol=4)

    plt.show()
```

```python
def compute_EC_values(configurations):
    """
    Computes energy consumption (EC) values for given vehicle configurations.

    Parameters:
    - configurations (list of lists): Each sublist contains the category names for:
        [body, engine, transmission, path, driver]

    Returns:
    - EC_values (list): Computed EC values for each configuration.
    """
    EC_values = []
    for config in configurations:
        if len(config)==6:
            body, engine, trans, path, driver, battery = config
            EC_values.append(model.EC_el(body_dict_EV[body], engine_dict_EV[engine], trans_dict_EV[trans], path_dict[path], driver_dict[driver], battery_dict_EV[battery])[0]/8.9)
        else:
            body, engine, trans, path, driver = config
            EC_values.append(model.EC_th(body_dict[body], engine_dict[engine], trans_dict[trans], path_dict[path], driver_dict[driver])[0])
    return EC_values


def main():
    """
    Main function to compute energy consumption uncertainty and visualize it.
    """
    # Define vehicle configurations
    light_min = ['Mini', 'Small_min', 'Manual - FWD', 'NEDC', 'Extra Eco']
    light_average_EU = ['Compact', 'Small_EU', 'Average_EU', 'EU_mix', 'Average_EU']
    light_average_US = ['Lower Med.', 'Small_US', 'Average_US', 'US_mix', 'Average_US']
    light_max = ['Lower Med.', 'Small_max', 'Manual - AWD', 'US06', 'Extra Aggres.']

    regular_min = ['Lower Med.', 'Medium_min', 'Manual - FWD', 'NEDC', 'Extra Eco']
    regular_average_EU = ['Upper Med.', 'Average_EU', 'Average_EU', 'EU_mix', 'Average_EU']
    regular_average_US = ['SUV', 'Average_US', 'Average_US', 'US_mix', 'Average_US']
    regular_max = ['SUV', 'Medium_max', 'Manual - AWD', 'US06', 'Extra Aggres.']

    heavy_min = ['SUV', 'Large_min', 'Manual - FWD', 'NEDC', 'Extra Eco']
    heavy_average_EU = ['SUV', 'Large_EU', 'Average_EU', 'EU_mix', 'Average_EU']
    heavy_average_US = ['Pick-ups', 'Large_US', 'Average_US', 'US_mix', 'Average_US']
    heavy_max = ['Pick-ups', 'Large_max', 'Manual - AWD', 'US06', 'Extra Aggres.']
    
    
    elec_min=['Mini','Extra Small','FWD','City - Mean','Extra Eco','Best']
    elec_EU=['Average_EU','Average_EU','Average_EU','EU_mix','Average_EU','Average']
    elec_US=['Average_US','Average_US','Average_US','US_mix','Average_US','Average']
    elec_max=['Pick-ups','Extra Large','RWD','US06','Extra Aggres.','Worst']


    # Reference Ecoinvent values
    ecoinvent = [6.71, 8.36, 9.84,19.9/8.9]

    # Organizing vehicle configurations into categories
    lighter = [light_min, regular_min, heavy_min,elec_min]
    heavier = [light_max, regular_max, heavy_max,elec_max]
    average_EU = [light_average_EU, regular_average_EU, heavy_average_EU,elec_EU]
    average_US = [light_average_US, regular_average_US, heavy_average_US,elec_US]

    # Compute EC values
    EC_min = compute_EC_values(lighter)
    EC_max = compute_EC_values(heavier)
    EC_av_EU = compute_EC_values(average_EU)
    EC_av_US = compute_EC_values(average_US)

    # Plot the results
    title = "Energy Consumption Uncertainty Analysis"
    comparison_uncertainty_box(ecoinvent, EC_min, EC_max, EC_av_EU, EC_av_US)


# Run the main function
if __name__ == "__main__":
    main()
```

# Application 3 - Lightweighting Case Study

```python
### initialisation of the dictionnaries with the pre-set configurations.

contributor=['Body', 'Engine', 'Drivetrain', 'Path', 'Driver','Battery']

body_dict=file_opening_dict('body.csv')
engine_dict=file_opening_dict('thermal_ecoinvent.csv')
trans_dict=file_opening_dict('transmission.csv')
path_dict=file_opening_dict('path.csv')
driver_dict=file_opening_dict('driver.csv')

body_dict_EV=file_opening_dict('body_EV.csv')
engine_dict_EV=file_opening_dict('electric.csv')
trans_dict_EV=file_opening_dict('trans_EV.csv')
battery_dict_EV=file_opening_dict('battery.csv')
```

```python
def list_MIEC(L):
    """
    Computes the Mass Influences on Energy Consumption (MIEC) with or without secondary effect (SE)
    for a given list of GV configurations.

    Parameters:
    - L (list of lists): Each sublist contains category names for:
        [body, engine, transmission, path, driver]

    Returns:
    - MIEC (list): Computed MIEC values for each configuration.
    - MIEC_SE (list): Computed MIEC_SE including secondary reduction effects (gear ratio)
    """
    
    MIEC, MIEC_SE = [], []

    for scenario in L:
        body_category, engine_category, trans_category, path_category, driver_category = scenario

        # Retrieve category-specific parameters
        body = body_dict[body_category]
        engine = engine_dict[engine_category]
        trans = trans_dict[trans_category]
        path = path_dict[path_category]
        driver = driver_dict[driver_category]

        # Compute MIEC and MIEC_SE
        piec_values = model.PIEC_th(body, engine, trans, path, driver)
        MIEC.append(piec_values[0])    
        MIEC_SE.append(piec_values[9]) 

    return MIEC, MIEC_SE
```

```python
def list_MIEC_EV(L):
    """
    Computes the Mass Influences on Energy Consumption (MIEC) with or without secondary effect (SE)
    for a given list of BEV configurations.

    Parameters:
    - L (list of lists): Each sublist contains category names for:
        [body, engine, transmission, path, driver,battery]

    Returns:
    - MIEC (list): Computed MIEC values for each configuration.
    - MIEC_SE (list): Computed MIEC_SE including secondary reduction effects (gear ratio)
    """
    
    MIEC, MIEC_SE = [], []

    for scenario in L:
        body_category, engine_category, trans_category, path_category, driver_category, battery_category = scenario

        # Retrieve category-specific parameters
        body = body_dict_EV[body_category]
        engine = engine_dict_EV[engine_category]
        trans = trans_dict_EV[trans_category]
        path = path_dict[path_category]
        driver = driver_dict[driver_category]
        battery = battery_dict_EV[battery_category]

        # Compute MIEC and MIEC_SE
        piec_values = model.PIEC_el(body, engine, trans, path, driver, battery)
        MIEC.append(piec_values[0])
        MIEC_SE.append(piec_values[8]) 

    return MIEC, MIEC_SE
```

```python
### The 4 configurations under study for GV ###

Reference=['Average_EU','Average_EU','Average_EU','EU_mix','Average_EU']
Small=['Compact','Small_EU','Manual - FWD','EU_mix','Average_EU']
City=['Average_EU','Average_EU','Manual - FWD','City - Mean','Average_EU']
US=['Average_US','Average_US','Average_US','US_mix','Average_US']

MIEC,MIEC_SE=list_MIEC([Reference,Small,City,US])

### The 4 configurations under study for BEV ###
Reference=['Average_EU','Average_EU','Average_EU','EU_mix','Average_EU','Average']
Small=['Compact','Small','FWD','EU_mix','Average_EU','Average']
City=['Average_EU','Average_EU','FWD','City - Mean','Average_EU','Average']
US=['Average_US','Average_US','Average_US','US_mix','Average_US','Average']

MIEC_EV,MIEC_EV_SE=list_MIEC_EV([Reference,Small,City,US])


### Quantification of the parameters introduced in the streamlined approach ###
petrol=4.37*3.6/43.5
elec_QC=0.0141
elec_US=0.4793
elec_CH=0.0333
elec_EU=0.3285

D_life=150000 #km
other_exhaust=6.08e-2 #kgCO2/(100kg.100km)
other_exhaust=11e-2 #kgCO2/(100kg.100km)
```

```python
def add_bicolor_ellipse(ax, center, width, height,c1,c2):
    """
    Fonction pour ajouter une ellipse bicolore (gauche-droite) à une figure existante.

    :param ax: Les axes de la figure où l'ellipse sera ajoutée.
    :param center: tuple (x, y), centre de l'ellipse.
    :param width: largeur de l'ellipse.
    :param height: hauteur de l'ellipse.
    """
    # Demi-ellipse gauche (couleur 1)
    theta_left = np.linspace(-np.pi/2, np.pi/2, 100)  # Moitié gauche de l'ellipse
    x_left = (width / 2) * np.cos(theta_left) + center[0]
    y_left = height * np.sin(theta_left) + center[1]
    ax.fill(x_left, y_left, color=c1, zorder=3)

    # Demi-ellipse droite (couleur 2)
    theta_right = np.linspace(np.pi/2, 3*np.pi/2, 100)  # Moitié droite de l'ellipse
    x_right = (width / 2) * np.cos(theta_right) + center[0]
    y_right = height * np.sin(theta_right) + center[1]
    ax.fill(x_right, y_right, color=c2, zorder=2)
    
    
def add_circle(ax, center, width, height, color='yellow'):
    """
    Adds a circle around a specific point or shape on the plot.
    
    Parameters:
    ax (matplotlib axis): The axis on which to draw the circle.
    center (tuple): The (x, y) coordinates for the center of the circle.
    radius (float): The radius of the circle.
    color (str): The color of the circle.
    linestyle (str): The style of the circle's line.
    linewidth (float): The width of the circle's line.
    """
    circle = Ellipse(center, width, height, color=color, fill=False, linewidth=1.5)
    ax.add_patch(circle)
```

```python
def Sobol (body,engine,trans,path,driver):
    EC_ref=model.EC_th(body,engine,trans,path,driver)[0]
    contributors=[body,engine,trans,path,driver]
    sobol=[]
    for k in range (len(contributors)):
        contributor=contributors[k]
        for i in range (len(contributor)):
            contributor[i]*=1.01
            EC=model.EC_th(*contributors)[0]
            sobol+=[np.abs((EC/EC_ref-1)*100)]
            contributor[i]/=1.01
    return sobol
```

```python
def Sobol_scenarios (scenarii):
    
    results=[]
    for scenario in scenarii:
        results+=[Sobol(body_dict[scenario[0]],engine_dict[scenario[1]],trans_dict[scenario[2]],path_dict[scenario[3]],driver_dict[scenario[4]])]
            
    results=np.array(results)
    return results
```

```python
def heat_maps (Scenarios,parameters_filtered=None,scenario_filtered=None):
    sobol_array=Sobol_scenarios(Scenarios)
    x_labels=['M_body','r0','Cd','A','Iw','Rw','transf','Pacc','M_eq','Cd_eq','P_max','ne','D','fmep0','p0','Q0','N_idle','cs','stop_start','ntr','a_tr','S','Ne','J3p','K1p','K2p','rate_acc','J0p','H','w','urban','dist','t_idle','B','mu_v','mu_a','mu_N','M_payload']
    y_labels=labels_cat
    
    if parameters_filtered is None:
        parameters_filtered = x_labels
    if scenario_filtered is None:
        scenario_filtered = y_labels


    # Find indices for the selected labels
    x_indices = [x_labels.index(label) for label in parameters_filtered if label in x_labels]
    y_indices = [y_labels.index(label) for label in scenario_filtered if label in y_labels]
    
    filtered_array = sobol_array[np.ix_(y_indices, x_indices)]
    #sobol_array=np.transpose(sobol_array)
    
    fig, ax = plt.subplots(figsize=(25,20))
    cax=ax.imshow(filtered_array, cmap='Reds')
    
    threshold = np.max(filtered_array) / 2.0
    for (i, j), val in np.ndenumerate(filtered_array):
        color = 'white' if val > threshold else 'black'
        ax.text(j, i, f'{val:.2f}', ha='center', va='center', color=color,fontsize=15)
    
    
    ax.set_xticks(np.arange(len(parameters_filtered)))
    ax.set_yticks(np.arange(len(scenario_filtered)))
    ax.set_xticklabels(parameters_filtered,fontsize=16)
    ax.set_yticklabels(scenario_filtered,fontsize=16)
    plt.xticks(rotation=45)
    fig.colorbar(cax, shrink=0.5)  # Ajoute une barre de couleur à côté de la heatmap

    # Affichage du graphique
    plt.show()
```

```python
param_filt=['M_body','r0','Cd','transf','Pacc','ne','D','fmep0','p0','ntr','Ne','J3p','K1p','K2p','J0p','B','mu_v','mu_a','mu_N']
heat_maps(Scenarios,param_filt)
```
